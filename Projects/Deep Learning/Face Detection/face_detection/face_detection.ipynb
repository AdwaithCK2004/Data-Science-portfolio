```python
import cv2
import matplotlib.pyplot as  plt
```

```python
img=cv2.imread('group_photo.jpg')
img
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
facedata=cv2.CascadeClassifier('haarcascade_frontalface_alt (1).xml')
facedata
```

```python
#face detect
faces=facedata.detectMultiScale(conv_img,minNeighbors=5)
faces
#- minNeighbors=5  requires at least 5 overlapping detections before confirming a face.
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(conv_img,pt1=(x,y),pt2=(x+w,y+h),color=(255,0,0),thickness=3)
plt.imshow(conv_img)    
```

```python
img2=cv2.imread('group_img2.jpg')
conv_img1=cv2.cvtColor(img2,cv2.COLOR_RGB2BGR)
plt.imshow(conv_img1)
```

```python
faces2=facedata.detectMultiScale(conv_img1,minNeighbors=5)
faces2
```

```python
for (x1,y1,w1,h1) in faces2:
    cv2.rectangle(conv_img1,pt1=(x1,y1),pt2=(x1+w1,y1+h1),color=(255,0,0),thickness=3)
plt.imshow(conv_img1)    
```
