```python
import pandas as pd
import numpy as np
import face_recognition as fr
import cv2
import os
```

```python
path='attendance_images'
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
names=[]
for i in mylist:
    image_path=os.path.join(path,i)
    current_img=cv2.imread(image_path)
    imgs.append(current_img)
    names.append(i.split('.')[0])
imgs    
```

```python
names
```

```python
def face_encoding(image):
    encodedlist=[]
    for img in image:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img)
        face_encoded=fr.face_encodings(img,face_in_frame)[0]
        encodedlist.append(face_encoded)
    return encodedlist    
```

```python
encoded_known_face=face_encoding(imgs)
encoded_known_face
```

```python
from datetime import datetime
import os

attendance_file1 = "attendance.csv"

# Create file if not exists
if not os.path.exists(attendance_file1):
    with open(attendance_file1, "w") as f:
        f.write("Name,Date,Time\n")

def mark_attendence1(name):
    date = datetime.now().strftime("%Y-%m-%d")   # only date
    now_time = datetime.now().strftime("%H:%M:%S")
    # check if already marked
    with open(attendance_file1, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if name in line and date in line:
                return   # already marked today

    # write new attendance
    with open(attendance_file1, 'a') as f:
        f.write(f"{name},{date},{now_time}\n")
    # print(f"Attendance marked for {name} on {date}"
```

```python
import pyttsx3
txt_sp=pyttsx3.init()
voices = txt_sp.getProperty('voices')
            #female voice
txt_sp.setProperty('voice', voices[1].id)
txt_sp.setProperty('volume',1.0)


video = cv2.VideoCapture(0)

while True:
    suc, img = video.read()

    img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    face_in_frame2 = fr.face_locations(img2)
    

    encoded_face = fr.face_encodings(img2,face_in_frame2)

    for face_loc, enc in zip(face_in_frame2,encoded_face):
        matches = fr.compare_faces(encoded_known_face,enc)
        # print(matches)
        face_distance=fr.face_distance(encoded_known_face,enc)
        # print(face_distance)
        
        match_index=np.argmin(face_distance)
        if matches[match_index]==True:
            ind=match_index
            name=names[match_index]
            print(name)
            print(ind)
            txt_sp.say(name)

            txt_sp.runAndWait()
            mark_attendence1(name)

#draw rectange
            y1,x2,y2,x1=face_loc
            cv2.rectangle(img,pt1=(x1,y1),pt2=(x2,y2),color=(0,0,255),thickness=3)
        cv2.imshow("Window",img)
        txt_sp.runAndWait()

    if cv2.waitKey(1) & 0xff == ord('q'):
        break

video.release()
cv2.destroyAllWindows()        
```
