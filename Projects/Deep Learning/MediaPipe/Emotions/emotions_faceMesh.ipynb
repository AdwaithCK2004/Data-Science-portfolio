```python
import mediapipe as mp
import cv2
import math
```

```python
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils
```

```python
face_mesh=mp_face_mesh.FaceMesh()
```

```python
video = cv2.VideoCapture(0)

while True:
    suc, img = video.read()
    if not suc:
        break

    h, w, _ = img.shape

    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = face_mesh.process(img1)

    if result.multi_face_landmarks:
        for x in result.multi_face_landmarks:
            mp_drawing.draw_landmarks(
                img,
                x,
                mp_face_mesh.FACEMESH_TESSELATION,
                mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=1, circle_radius=1),
                mp_drawing.DrawingSpec(color=(0, 0, 255), thickness=1)
            )

            # ---------------- MOUTH ----------------
            left_mouth = x.landmark[61]
            right_mouth = x.landmark[291]
            upper_lip = x.landmark[13]
            lower_lip = x.landmark[14]

            x1, y1 = int(left_mouth.x * w), int(left_mouth.y * h)
            x2, y2 = int(right_mouth.x * w), int(right_mouth.y * h)

            mouth_width = math.hypot(x2 - x1, y2 - y1)
            mouth_height = abs(int(upper_lip.y * h) - int(lower_lip.y * h))

            # ---------------- EYES (for crying) ----------------
            left_eye_top = x.landmark[159]
            left_eye_bottom = x.landmark[145]
            eye_height = abs(int(left_eye_top.y * h) - int(left_eye_bottom.y * h))

            # ---------------- EMOTION LOGIC ----------------
            if mouth_width > 60:
                emotion = "SMILING "
                color = (0, 255, 0)

            elif mouth_width < 45 and eye_height > 8:
                emotion = "SAD "
                color = (0, 0, 255)

            elif mouth_width < 40 and eye_height < 6:
                emotion = "CRYING "
                color = (0, 0, 255)

            else:
                emotion = "NEUTRAL "
                color = (0, 0, 255)

            cv2.putText(img, emotion, (30, 50),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

    cv2.imshow("image", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
